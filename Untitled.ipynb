```r
library(SingleCellExperiment)
library(Seurat)
library(scmap)
library(scAlign)
```

```r
load_files <- function(directory) {
    files <- list.files(directory, full.names=TRUE)
    data.list <- lapply(files, function(x) {as.data.frame(read.csv(x, row.names=1))})
    return(data.list)
}
```

```r
dataset <- "pancreas"

print("Loading provided files")
data.directory <- file.path("data/normalized/", dataset, "/data")
data.list <- load_files(data.directory)
files <- list.files(data.directory, full.names=FALSE)
files <- lapply(files, function(x) sub(pattern = "(.*)\\..*$", replacement = "\\1", basename(x)))
```

```r
dataset.list <- lapply(data.list, CreateSeuratObject)

for (i in 1:length(dataset.list)) {
    dataset.list[[i]] <- ScaleData(dataset.list[[i]])
    dataset.list[[i]] <- FindVariableFeatures(dataset.list[[i]])
}
```

```r
pancreas.sce.list = lapply(dataset.list,
                           function(seurat.obj){
                             SingleCellExperiment(assays = list(counts = seurat.obj@assays$RNA@counts,
                                                                logcounts = seurat.obj@assays$RNA@counts,
                                                                scale.data = seurat.obj@assays$RNA@scale.data),
                                                  colData = seurat.obj@meta.data)
                            })
```

```r
genes.use <- rownames(dataset.list[[1]])
```

```r
scAlignPancreas = scAlignCreateObject(sce.objects = pancreas.sce.list,
                                      genes.use = genes.use,
                                      cca.reduce=T,
                                      ccs.compute=10,
                                      project.name = "scAlign_Pancreatic_Islet")
```

```r
scAlignPancreas = scAlignMulti(scAlignPancreas,
                        options=scAlignOptions(steps=100,
                                               log.every=1,
                                               batch.size=300,
                                               perplexity=30,
                                               norm=TRUE,
                                               batch.norm.layer=FALSE,
                                               architecture="large",  ## 3 layer neural network
                                               num.dim=64),            ## Number of latent dimensions
                        encoder.data="counts",
                        decoder.data="scale.data",
                        supervised='none',
                        run.encoder=TRUE,
                        run.decoder=FALSE,
                        log.dir=file.path('./tmp2'),
                        device="GPU")
```
